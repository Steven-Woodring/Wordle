```python
# Import dependencies
import pandas as pd

# Read in words and guess_scores lists from Guess_Scores.ipynb
%store -r words
%store -r guess_scores
```

```python
# Make list of top 10 first guesses from guess_scores list
top_first_guesses = [guess['Word'] for guess in guess_scores[:10]]
top_first_guesses
```

```python
# Find index of most recent Wordle solution within words list (solutions are at the beginning of the list)
current_solution_idx = words.index("lilac")
current_solution_idx
```

```python
# Using index of most recent solution, get list of all past solutions
past_solutions = words[:(current_solution_idx + 1)]
past_solutions
```

```python
# Create function that will update colors of letters in a series of guesses based on the solution
def colorize_letters(guess, solution, grey_letters, yellow_letters_dict, green_letters_dict, first_guess=False):
    
    for letter_pos in range(5):
            
        letter = guess[letter_pos]

        if letter not in solution:
            grey_letters += letter
            
        elif letter == solution[letter_pos]:
            if first_guess == True:
                green_letters_dict[letter] = [letter_pos]
            else:
                if letter in green_letters_dict:
                    if letter_pos not in green_letters_dict[letter]:
                        green_letters_dict[letter].append(letter_pos)
                else:
                    green_letters_dict[letter] = [letter_pos]
                    
        else:
            if first_guess == True:
                yellow_letters_dict[letter] = [letter_pos]
            else:
                if letter in yellow_letters_dict:
                    yellow_letters_dict[letter].append(letter_pos)
                else:
                    yellow_letters_dict[letter] = [letter_pos]
            
    
    return grey_letters,yellow_letters_dict,green_letters_dict
```

```python
# Create function that will find the best next guess based on the information obtained from colorize_letters function
def find_next_guess(grey_letters, yellow_letters_dict, green_letters_dict, words_series=[guess['Word'] for guess in guess_scores]):
    
    for word in words_series:
            
        word_fits_grey = True
        word_fits_yellow = True
        word_fits_green = True

        for letter in grey_letters:
            if letter not in word:
                word_fits_grey = True
            else:
                word_fits_grey = False
                break

        for letter in yellow_letters_dict.keys():
            if letter not in word:
                word_fits_yellow = False
                break
            else:
                for idx in yellow_letters_dict[letter]:
                    if word[idx] != letter:
                        word_fits_yellow = True
                    else:
                        word_fits_yellow = False
                        break
                if word_fits_yellow == False:
                    break

        for letter in green_letters_dict.keys():
            for idx in green_letters_dict[letter]:
            #idx = green_letters_dict[letter]
                if word[idx] == letter:
                    word_fits_green = True
                else:
                    word_fits_green = False
                    break
            if word_fits_green == False:
                break

        if word_fits_grey & word_fits_yellow & word_fits_green:
            return word
```

```python
top_guess_results = {
    "aeros": [],
    "soare": [],
    "arose": [],
    "reais": [],
    "raise": [],
    "serai": [],
    "arise": [],
    "aesir": [],
    "lares": [],
    "rales": []
}


for solution in past_solutions:
    
    for guess in top_first_guesses:
        
        guess_counter = 1
        
        if guess == solution:
            break
        
        grey_letters,yellow_letters_dict,green_letters_dict = colorize_letters(guess, solution, "", {}, {}, True)
        
        next_guess = ""
        while next_guess != solution:

            guess_counter += 1
            next_guess = find_next_guess(grey_letters, yellow_letters_dict, green_letters_dict)
            grey_letters,yellow_letters_dict,green_letters_dict = colorize_letters(next_guess, solution, grey_letters, yellow_letters_dict, green_letters_dict)
            
            continue
        
        top_guess_results[guess].append(guess_counter)

top_guess_results
```

```python
top_guesses_metrics = []

for guess in top_guess_results:
    
    results = top_guess_results[guess]
    
    attempts = len(past_solutions)
    wins = len([result for result in results if result <= 6])
    win_pct = "{:.2%}".format(wins / attempts)
    
    avg_result = round((sum(results) / attempts), 2)
    
    wins_on_try1 = len([result for result in results if result == 1])
    wins_on_try2 = len([result for result in results if result == 2])
    wins_on_try3 = len([result for result in results if result == 3])
    wins_on_try4 = len([result for result in results if result == 4])
    wins_on_try5 = len([result for result in results if result == 5])
    wins_on_try6 = len([result for result in results if result == 6])
    
    guess_metrics = {
        'First_Guess': guess,
        'Win_pct': win_pct,
        'Avg_Result': avg_result,
        'Wins_On_Try1': wins_on_try1,
        'Wins_On_Try2': wins_on_try2,
        'Wins_On_Try3': wins_on_try3,
        'Wins_On_Try4': wins_on_try4,
        'Wins_On_Try5': wins_on_try5,
        'Wins_On_Try6': wins_on_try6
    }
    
    top_guesses_metrics.append(guess_metrics)
    
top_guesses_metrics

metrics_df = pd.DataFrame(top_guesses_metrics)
metrics_df.sort_values(by=["Win_pct", "Avg_Result"], ascending=False, inplace=True)
```

```python
# Sort top_guesses_metrics list by Avg_Result
top_guesses_metrics.sort(key=lambda x: x['Avg_Result'])
# Sort top_guesses_metrics list by Win_pct, descending
top_guesses_metrics.sort(key=lambda x: x['Win_pct'], reverse=True)

# Add a rank key/value pair to beginning of every guess dict within the list
for i in range(len(top_guesses_metrics)):
    rank_dict = {'Rank': i + 1}
    top_guesses_metrics[i] = {**rank_dict, **top_guesses_metrics[i]}

top_guesses_metrics
```
